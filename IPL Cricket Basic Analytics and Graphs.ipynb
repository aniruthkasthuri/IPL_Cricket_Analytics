```python
import pandas as panda
import csv
import numpy as np
import matplotlib.pyplot as plt
import pylab
import matplotlib

# @Source for CSV Files - Kaggle.com and Harsha Vardhan
# Link for Source - https://www.kaggle.com/harsha547/indian-premier-league-csv-dataset
```

```python
ifile = open('./ipl/Match.csv')
reader = csv.reader(ifile)
x = []
pos = 0
for row in reader:
    x.append(row)
    pos+=1
y = 0
bywickets = 0
byruns = 0
numwicks = []
numruns = []
wicks = []

while (y < len(x)):
    if x[y][11] == 'by wickets':
        wicks.append(bywickets)
        numwicks.append(int(x[y][12]))
        bywickets += 1
    else:
        byruns +=1
        numruns.append(x[y][12])
    y+=1
numwicks.sort(key=int)

#print(numwicks)
plt.figure(figsize=(20,10))
plt.scatter(wicks,numwicks)
plt.axhline(y=np.mean(numwicks), color='r', linestyle='-',label='Mean @ 6.32')
plt.axhline(y=np.mean(numwicks)-np.std(numwicks), color='green', linestyle='-',label='1 Std Below @ 4.53')
plt.axhline(y=np.mean(numwicks)+np.std(numwicks), color='orange', linestyle='-',label='1 Std Above @ 8.11')
pylab.legend(loc='upper left')

plt.title('Relationship Between Number of IPL Cricket Matches Played and Number of Wickets Won By')
plt.xlabel('Number of IPL Cricket Matches Played')
plt.ylabel('Number of Wickets Won By')

plt.show()

```
